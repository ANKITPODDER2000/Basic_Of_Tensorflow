```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Homedata.csv")
data.head()
```

```python
plt.scatter(data['Size'],data['Price'])
plt.xlabel("Size")
plt.ylabel("Price")
```

```python
data['Price'] = data['Price'].apply(abs)
```

```python
plt.scatter(data['Size'],data['Price'])
plt.xlabel("Size")
plt.ylabel("Price")
```

```python
X = np.array(data['Size'])
```

```python
X = X.reshape((X.size,1))
X = np.insert(X,0,1,1)
y = np.array(data['Price']).reshape((data['Price'].size , 1))
print(X.shape)
print(y.shape)
```

```python
X = tf.Variable(X)
```

```python
y = tf.Variable(y)
```

```python
theta = tf.Variable([[0.1],[0.2]],dtype=tf.float64)
```

```python
print("Shape of X : ",X.shape,"\nShape of theta is : ",theta.shape)
```

```python
def hypothesis(X,theta):
    return tf.matmul(X,theta)
```

```python
def loss_function(X,theta,y):
    prediction = hypothesis(X,theta)
    return tf.keras.losses.mse(y,prediction)
```

```python
#loss_function(X,theta,y)
```

```python
opt = tf.keras.optimizers.Adam()
```

```python
for j in range(20000):
    opt.minimize(lambda: loss_function(X,theta,y), var_list=[theta])
    print(theta)
```
